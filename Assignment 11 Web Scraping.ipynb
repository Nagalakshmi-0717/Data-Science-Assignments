### Text Preprocessing

```python
import numpy as np # linear algebra
import pandas as pd
import string # special operations on strings
import spacy # language models

from matplotlib.pyplot import imread
from matplotlib import pyplot as plt
from wordcloud import WordCloud
%matplotlib inline
```

```python
review=[x.strip() for x in Data_1]# remove both the leading and the trailing characters
review=[x for x in review if x]# removes empty strings, because they are considered in Python as False
review
```

```python
review_text=' '.join(review)
review_text
```

```python
#Remove Punctuation
no_punc_text= review_text.translate(str.maketrans('', '', string.punctuation))
no_punc_text
```

```python
#Tokenization
from nltk.tokenize import word_tokenize
review_token= word_tokenize(no_punc_text)
print(review_token[100:200])
```

```python
len(review_token)
```

```python
#Remove stopwords
import nltk
from nltk.corpus import stopwords
nltk.download('punkt')
nltk.download('stopwords')

my_stop_words = stopwords.words('english')
my_stop_words.append('the')
my_stop_words.append('I')
my_stop_words.append('\n')
my_stop_words.append('im')
my_stop_words.append('now')
my_stop_words.append('please')
my_stop_words.append('want')
no_stop_token= [word for word in review_token if not word in my_stop_words]
print(no_stop_token[0:40])
```

```python
#Noramalize the data
lower_words = [x.lower() for x in no_stop_token]
print(lower_words[100:200])
```

```python
# Lemmatization
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
wnl = WordNetLemmatizer()
lemma_text=[wnl.lemmatize(text) for text in lower_words ]
```

```python
clean_text=' '.join(lemma_text)
clean_text
```

### Word Cloud

```python
# Define a function to plot word cloud
def plot_cloud(cloud):
    plt.figure(figsize=(40,30))
    plt.imshow(cloud)
    plt.axis('off')
cloud= WordCloud(width=3000,height=2000,background_color='black',max_words=50, colormap='Set1').generate(clean_text)
plot_cloud(cloud)
```

```python
##Part Of Speech Tagging
nlp = spacy.load('en_core_web_sm')

one_block = clean_text
doc_block = nlp(one_block)
spacy.displacy.render(doc_block, style='ent', jupyter=True)
```

```python
for token in doc_block[100:200]:
    print(token,token.pos_)  
```

```python
# Filtering the nouns and verbs only
nouns_verbs=[token.text for token in doc_block if token.pos_ in ('NOUN','VERB')]
print(nouns_verbs[100:200])
```

```python
# Counting the noun & verb tokens
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()

X=cv.fit_transform(nouns_verbs)
sum_words=X.sum(axis=0)

words_freq=[(word,sum_words[0,idx]) for word,idx in cv.vocabulary_.items()]
words_freq=sorted(words_freq, key=lambda x: x[1], reverse=True)

wd_df=pd.DataFrame(words_freq)
wd_df.columns=['word','count']
wd_df[0:10] # viewing top ten results
```

```python
# Visualizing results (Barchart for top 10 nouns + verbs)
wd_df[0:10].plot.bar(x='word',figsize=(12,8),title='Top 10 nouns and verbs');
```

### Emotion Mining

```python
#Sentiment analysis
afinn = pd.read_csv('Afinn.csv', sep=',', encoding='latin-1')
afinn.shape
```

```python
afinn.head()
```

```python
from nltk import tokenize
reviews=tokenize.sent_tokenize(' '.join(Data_1))
reviews
```

```python
sent_df=pd.DataFrame(reviews,columns=['Review'])
sent_df
```

```python
affinity_scores=afinn.set_index('word')['value'].to_dict()
affinity_scores
```

```python
# Custom function: score each word in a sentence in lemmatised form, but calculate the score for the whole original sentence
nlp=spacy.load('en_core_web_sm')
sentiment_lexicon=affinity_scores

def calculate_sentiment(text:str=None):
    sent_score=0
    if text:
        sentence=nlp(text)
        for word in sentence:
            sent_score+=sentiment_lexicon.get(word.lemma_,0)
    return sent_score
```

```python
# manual testing
calculate_sentiment(text='good')
```

```python
# Calculating sentiment value for each sentence
sent_df['sentiment_value']=sent_df['Review'].apply(calculate_sentiment)
sent_df['sentiment_value']
```

```python
# how many words are there in a sentence?
sent_df['word_count']=sent_df['Review'].str.split().apply(len)
sent_df['word_count']
```

```python
sent_df.sort_values(by='sentiment_value')
```

```python
# Sentiment score of the whole review
sent_df['sentiment_value'].describe()
```

```python
# negative sentiment score of the whole review
sent_df[sent_df['sentiment_value']<=0]
```

```python
# positive sentiment score of the whole review
sent_df[sent_df['sentiment_value']>0]
```
